```python
from pyhive import hive

def create_employees_table():
    connection = hive.Connection(host='localhost', port=10000, username='your_username')
    cursor = connection.cursor()

    # Define the HiveQL query to create the Employees table
    create_table_query = """
    CREATE TABLE IF NOT EXISTS Employees (
        id INT,
        name STRING,
        salary FLOAT
    )
    """

    try:
        # Execute the HiveQL query to create the table
        cursor.execute(create_table_query)
        print("Table 'Employees' created successfully.")
    except Exception as e:
        print("Error creating table:", str(e))

    # Close the cursor and connection
    cursor.close()
    connection.close()

# Example usage:
if __name__ == "__main__":
    create_employees_table()
```

```python
from pyhive import hive

def retrieve_customers_greater_than_30():
    connection = hive.Connection(host='localhost', port=10000, username='your_username')
    cursor = connection.cursor()

    # Define the HiveQL query to retrieve records where age is greater than 30
    retrieve_query = """
    SELECT *
    FROM Customers
    WHERE age > 30
    """

    try:
        # Execute the HiveQL query to retrieve the records
        cursor.execute(retrieve_query)

        # Fetch all the records from the result set
        result = cursor.fetchall()

        # Print the retrieved records
        print("Customers with age greater than 30:")
        for record in result:
            print(record)

    except Exception as e:
        print("Error retrieving records:", str(e))

    # Close the cursor and connection
    cursor.close()
    connection.close()

# Example usage:
if __name__ == "__main__":
    retrieve_customers_greater_than_30()
```

```python
from pyhive import hive

def sort_logs_by_timestamp_desc():
    connection = hive.Connection(host='localhost', port=10000, username='your_username')
    cursor = connection.cursor()

    # Define the HiveQL query to sort records by timestamp in descending order
    sort_query = """
    SELECT *
    FROM Logs
    ORDER BY timestamp DESC
    """

    try:
        # Execute the HiveQL query to sort the records
        cursor.execute(sort_query)

        # Fetch all the sorted records from the result set
        sorted_logs = cursor.fetchall()

        # Print the sorted records
        print("Sorted Logs by Timestamp (Descending Order):")
        for record in sorted_logs:
            print(record)

    except Exception as e:
        print("Error sorting records:", str(e))

    # Close the cursor and connection
    cursor.close()
    connection.close()

# Example usage:
if __name__ == "__main__":
    sort_logs_by_timestamp_desc()
```

```python
from pyhive import hive

def retrieve_all_products():
    connection = hive.Connection(
        host='localhost',          
        port=10000,                
        username='your_username'   
    )
    cursor = connection.cursor()

    # Define the HiveQL query to retrieve all records from the "Products" table
    retrieve_query = "SELECT * FROM Products"

    try:
        # Execute the HiveQL query to retrieve the records
        cursor.execute(retrieve_query)

        # Fetch all the records from the result set
        all_products = cursor.fetchall()

        # Print the retrieved records
        print("All Products:")
        for record in all_products:
            print(record)

    except Exception as e:
        print("Error retrieving records:", str(e))

    # Close the cursor and connection
    cursor.close()
    connection.close()

# Example usage:
if __name__ == "__main__":
    retrieve_all_products()
```

```python
from pyhive import hive

def calculate_average_salary():
    connection = hive.Connection(
        host='localhost',          # Replace with the hostname or IP of your Hive server
        port=10000,                # Replace with the Hive server port
        username='your_username'   # Replace with your Hive username (if needed)
    )
    cursor = connection.cursor()

    # Define the HiveQL query to calculate the average salary from the "Employees" table
    average_salary_query = "SELECT AVG(salary) AS average_salary FROM Employees"

    try:
        # Execute the HiveQL query to calculate the average salary
        cursor.execute(average_salary_query)

        # Fetch the result
        result = cursor.fetchone()

        # Print the average salary
        average_salary = result[0]
        print("Average Salary of Employees:", average_salary)

    except Exception as e:
        print("Error calculating average salary:", str(e))

    # Close the cursor and connection
    cursor.close()
    connection.close()

# Example usage:
if __name__ == "__main__":
    calculate_average_salary()
```
